## Continuity and Differentiation
### Limits and Continuity

**Let:** $f:D→ℝ$ on $D⊆ℝ^d$. $f$ has a limit $L∈ℝ$ as $x$ approaches $a$ if for any $ϵ>0$ there exist a $δ>0$ such that $|f(x)-L|<ϵ$ $∀ x∈D∩B_δ(a)$/{a} \
**Limits of a function:** $lim_{x->a}f(x)=L$ \
**Continuous function:** $lim_{x->a}f(x)=f(a)$ \
**Extremum:** $f$ obtains a maximum value M at $x_M$ if $f(x_M)=M$ and $M≥f(x)$ $∀x∈D$. The same applies for a minimum! $f$ obtains a maximum m at $x_m$ if $f(x_m)=m$ and $m≤f(x)$ $∀x∈D$
\
**Lemma: Composition of Continuous Functions** - Let $f:D_1->ℝ^m$ and $g:D_2->ℝ^p$ where $D_1⊆ℝ^d$ and $D_2⊆ℝ^m$. Asumming $f$ is continuous at $x_0$ and g is continuous at $f(x_0)$ then the composition of g and f is continuous at $x_0$.\

```python
# Necessary imports
import numpy as np
import matplotlib.pyplot as plt

# Continuous function
x = np.linspace(-10,10,200)
c_y = lambda x: x**2

# Discontinuous
dc_y = lambda x: 1/(x-5)

# plot function
plt.figure(figsize=(10,8))
plt.plot(x, c_y(x), 'k', label = 'Continuous')
plt.plot(x, dc_y(x), 'r', label = 'Discontinuous')
plt.grid()
plt.title('Continuous and Discontinuous Functions')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

### Differentiation

**Single Variable Differentiation:** Let $f:D\rightarrowℝ$ with $D⊆ℝ$ and $x_0∈D$ is an interior point of D. \
**First Order Derivative:** $f'(x_0)=\frac{df(x_0)}{dx}=lim_{h→0}\frac{f(x_0+h)-f(x_0)}{h}$ \
**Second Order Derivative:** $f''(x_0)=\frac{d^2f(x_0)}{dx^2}=lim_{h→0}\frac{f'(x_0+h)-f'(x_0)}{h}$ \
**Mean Value Theorem:** Let $a<c<b$ then $f(b) = f(a) + (b-a)f'(c)$ or in other words $\frac{f(b)-f(a)}{b-a}=f'(c)$ \
\
**Multi Variable Differentiation:** Let $f:D→ℝ$ with $D⊆ℝ^d$ and $x_0∈D$ with $e_i∈ℝ^d$ is the i-th basis vector\
**Partial Derivative:** $\frac{∂f(x_0)}{∂x_i} = lim_{h→0}\frac{f(x_0+he_i)-f(x_0)}{h}$ \
**Jacobian:** Given $f:D→ℝ$ then the Jacobian is $J_f(x_0)=∇f(x_0)^T$ \
**Gradient:** $∇f(x_0) = (\frac{∂f(x_0)}{∂x_1},...,\frac{∂f(x_0)}{∂x_d})$ \
**Directional Derivative:** $\frac{∂f(x_0)}{∂v}=lim_{h→0}\frac{f(x_0+hv)-f(x_0)}{h}$\
**Directional Derivative from Gradient:** $\frac{∂f(x_0)}{∂v}=J_f(x_0)v=∇f(x_0)^Tv$ with $v$ being a unit vector \
**Second Partial Derivative:** $\frac{∂^2f(x_0)}{∂x_j∂x_i} = lim_{h→0}\frac{\partial f(x_0+he_j)/∂x_i- \partial f(x_0)/∂x_i}{h}$

```python
# Necessary imports
import numpy as np
import matplotlib.pyplot as plt

# Single Variable differentiation
x = np.linspace(-10,10,200)

# original function and then derivatives
f = lambda x: -5*(x**3) + 10*(x**2) - 3*x +10
fp = lambda x: -15*(x**2) + 20*x -3
fpp = lambda x: -30*x + 20
fppp = lambda x: -30 + (x*0)

# clear before plotting
plt.clf()

# plot derivatives
plt.figure(figsize=(10,8))
plt.plot(x, f(x), 'k', label = 'Original')
plt.plot(x, fp(x), 'b', label = 'First Order')
plt.plot(x, fpp(x), 'r', label = 'Second Order')
plt.plot(x, fppp(x), 'g', label = 'Third Order')
plt.grid()
plt.title('Derivatives of a Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Taylor Theorem

### Theory
**Taylor's theorem** determines an expansion of a function in an infinite series of polynomials around a given point. \
**Let -** $f: D → ℝ$ with $D ⊆ ℝ$. $f$ has $m$ continuous derivative on $[a,b]$
\
$f(b)=f(a)+(b-a)f'(a)+\frac{1}{2}(b-a)^2f''(a)+...+\frac{(b-a)^{m-1}}{(m-1)!}f^{m-1}(a)+R_m$ \
\
$R_m=\frac{(b-a)^m}{(m)!}f^m(a+Θ(b-a))$ for some $0 < Θ < 1$. \

### Applied

**Taylor Series Exapansion :** \
$f(x)=∑_{n=0}^{∞}\frac{f^n(a)(x-a)^n}{n!}$ with $f^n$ is the $n$th derivative \
\
**Example Problem:**\
$f(x)=cos(x)$ \
Taylor series expansion yields: \
$f(x) = 1- \frac{x^2}{2!}+\frac{x^4}{4!}$ \
In sum notation: \
$f(x) = Σ_{n=0}^{∞}\frac{(-1)^2}{(2n)!}x^{2n}$ \
Now we can use python below to generate approximations and get closer to the original function and compare to the analytic solution.


```python
# necessary imports
import numpy as np
import matplotlib.pyplot as plt

# clear before any plotting
plt.clf()

# functions to approximate
x = np.linspace(- np.pi, np.pi, 200)
y = np.zeros(len(x))

# Taylor series labels
labels = ['Zero Order', 'Second Order', 'Fourth Order', 'Sixth Order']

# Ploting approximations
plt.figure(figsize=(10,8))
for n, label in zip(range(4), labels):
  y = y + ((-1)**n * (x)**(2*n)) / np.math.factorial(2*n)
  plt.plot(x,y, label = label)

plt.plot(x, np.cos(x), 'k', label = 'Analytic')
plt.grid()
plt.title('Taylor Series Approximations of f(x)=cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## Optimization with Gradient Descent

### Theory
 **Gradient Descent** is an algorithm that works for functions which are differentiable. The algorithm works by moving in the direction which leads to smallet and smaller outputs of the optimized function. In short the algorithm is trying to find the minimum of a given function. A common application of Gradient descent is in least-squares regression models. In least squares regression the **mean square error** accounts for the loss function is therefore the optimization metric. The gradient descent algorithm adjust the objective function in a manner which reduces the loss until a local minima is reached. This approach boost the accuracy of machine learning models which rely on mean square error and root mean squared error for model fitness. \
 **Let -** $f:ℝ^d→ℝ$ with unconstrained minimization problems in form $min_{x ∈ ℝ^d}||Ax-b||^2$. \
 \
 **Steepest Descent** is one approach to find smaller values of $f$ by iteratively adjust the function. \
 $\frac{∂f(x_0)}{∂v}≥\frac{∂f(x_0)}{\partial v^*}$ where $v^*=\frac{∇f(x_0)}{||∇f(x_0)||}$
 \
 ### Application
 Observe an example of **Gradient Descent** being used to find the weight w that minimizes the loss function. In this example the loss function is: \
 $L(w)= 5x^2 -5x$ \
 $L'(w) = 10x - 5$

```python
# necessary imports
import numpy as np
import matplotlib.pyplot as plt

# clear before any plotting
plt.clf()

# Gradient Descent function - steepest descent
def gradient_descent(start_guess, start_guess_prev):
  w = start_guess
  w_prev = start_guess_prev
  w_list = [start_guess_prev, start_guess]
  stepSize = 0.1
  stop_accuracy = 0.005
  L = lambda w: 5*(w**2) - 5*w
  Lp = lambda w: 10*w - 5

  while abs(w-w_prev) > stop_accuracy:
    w_list.append(w)
    w_prev = w
    w = w_prev - stepSize*Lp(w_prev)

  return w, w_prev, w_list

weight, weight_prev, weights_list = gradient_descent(-10,-9.9)
L = lambda w: 5*(w**2) - 5*w
Lp = lambda w: 10*w - 5

# Demonstrate how Gradient Descent slowly adjust next guess based on past
# information
loss_improvement = list()
for weight in weights_list:
  loss_improvement.append(L(weight))

print('Ending Results: w =' + str(weight) + ', L(w) = ' +  str(L(weight)) + ', Lp(w) = ' + str(Lp(weight)))

# display x values
x = np.linspace(-10, 10, 200)

plt.figure(figsize=(10,8))
plt.plot(x, L(x), 'k', label = 'Loss Function')
plt.plot(x, Lp(x), 'b', label = 'Derivative Loss')
plt.plot(weight, L(weight), 'ro', label = 'Minimum Loss')
plt.plot(weight, Lp(weight), 'go', label = 'Minimum Loss Deriv')
plt.plot(weights_list, loss_improvement, 'y', label = 'Descent Guesses')
plt.grid()
plt.title('Gradient Descent Example')
plt.xlabel('x')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
